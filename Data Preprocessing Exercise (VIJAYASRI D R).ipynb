```python
titanic['Embarked']=titanic['Embarked'].map({'S':1,'C':2,'Q':3})
```

```python
titanic['Cabin']=np.where(titanic['Cabin'].isnull(),"Unknown",titanic['Cabin'])
```

```python
#impute the null values
titanic.Survived=titanic.Survived.fillna(titanic.Survived.mean())
titanic.Age=titanic.Age.fillna(titanic.Age.median())
titanic.Fare=titanic.Fare.fillna(titanic.Fare.mean())
titanic.Embarked=titanic.Embarked.fillna(titanic.Embarked.mean())
```

```python
titanic=titanic.drop(['Name'],axis=1)
titanic=titanic.drop(['Ticket'],axis=1)
titanic=titanic.drop(['Cabin'],axis=1)
```

```python
#One Hot Encoding

from sklearn.compose import ColumnTransformer #it is used for performing both OHE & LE in single step
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([('on',OneHotEncoder(),[3])],remainder='passthrough')
titanic=ct.fit_transform(titanic)
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
titanic=sc.fit_transform(titanic)
```
